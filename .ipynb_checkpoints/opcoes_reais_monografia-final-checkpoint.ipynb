# Notebook de estimação dos resultados

## Preparação do Ambiente

```python
import pandas as pd
import numpy as np

from scipy.optimize import fsolve, newton, root, bisect, curve_fit, broyden1

import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

## Definição dos parâmetros já estimados/definidos

```python
r = 0.03 # taxa de juros real ( 7%(SELIC/24) - 4%(IGP-M/24) = 3% (Boletim Focus 22/11/2021))

p_zero = 93.34 # Preço Inicial da Energia Elétrica
drift_p = 0.00664 # Drift da Energia Elétrica #0.664%
sigma_p = 0.0292 # Desvio-padrão da Energia Elétrica #2.92%

# I_eolica = 4995000
# c_eolica = 6.326
# i_eolica = I_eolica / c_eolica # 90.137
drift_eolica = 0.00971
sigma_eolica = 0.04316
rho_eolica = 0.1998

# I_solar = 2442000
# c_solar = 1.518
# i_solar = I_solar / c_solar # 183.641
drift_solar = 0.00971
sigma_solar = 0.04316
rho_solar = 0.1998

# I_gas = 1111000
# c_gas = 20.376
# i_gas = I_gas / c_gas # 6.224
drift_gas = 0.0051
sigma_gas = 0.1189
rho_gas = -0.0887

# I_carvao = 3289000
# c_carvao = 22.619
# i_carvao = I_carvao / c_carvao #16.599
drift_carvao = -0.003
sigma_carvao = 0.0748
rho_carvao = -0.0549 
```

## Definição das funções para lambda, constante B e razão crítica

### Lambda (parâmetro EDO)

```python
def edo_eolica(lamb):
    return (1/2)*((sigma_p ** 2) - (2 * rho_eolica * sigma_p * sigma_eolica) + (sigma_eolica ** 2)) * lamb * (lamb-1) + (drift_p - drift_eolica) * lamb - r

def edo_solar(lamb):
    return (1/2)*(sigma_p ** 2 - 2 * rho_solar * sigma_p * sigma_solar + sigma_solar ** 2) * lamb * (lamb-1) + (drift_p - drift_solar) * lamb - r

def edo_gas(lamb):
    return (1/2)*(sigma_p ** 2 - 2 * rho_gas * sigma_p * sigma_gas + sigma_gas ** 2) * lamb * (lamb-1) + (drift_p - drift_gas) * lamb - r

def edo_carvao(lamb):
    return (1/2)*(sigma_p ** 2 - 2 * rho_carvao * sigma_p * sigma_carvao + sigma_carvao ** 2) * lamb * (lamb-1) + (drift_p - drift_carvao) * lamb - r
```

### Constante B e Razão Crítica

```python
def sistema_eolica(x, i_eolica): # x = (x, B)
    y = np.zeros(np.size(x))
    y[0] = x[1] * (x[0] ** lamb_eolica) + (x[0] / (r - drift_p + drift_eolica)) - (1 / r) - i_eolica 
    y[1] = x[1] * lamb_eolica * (x[0] ** (lamb_eolica - 1)) + (1 / (r - drift_p + drift_eolica))
    return y

def sistema_solar(x): # x = (x, B)
    y = np.zeros(np.size(x))
    y[0] = x[1] * (x[0] ** lamb_solar) + (x[0] / (r - drift_p + drift_solar)) - (1 / r) - i_solar # primeira equação
    y[1] = x[1] * lamb_solar * (x[0] ** (lamb_solar - 1)) + (1 / (r - drift_p + drift_solar)) # segunda equação
    return y

def sistema_gas(x): # x = (x, B)
    y = np.zeros(np.size(x))
    y[0] = x[1] * (x[0] ** lamb_gas) + (x[0] / (r - drift_p + drift_gas)) - (1 / r) - i_gas # primeira equação
    y[1] = x[1] * lamb_gas * (x[0] ** (lamb_gas - 1)) + (1 / (r - drift_p + drift_gas)) # segunda equação
    return y

def sistema_carvao(x): # x = (x, B)
    y = np.zeros(np.size(x))
    y[0] = x[1] * (x[0] ** lamb_carvao) + (x[0] / (r - drift_p + drift_carvao)) - (1 / r) - i_carvao # primeira equação
    y[1] = x[1] * lamb_carvao * (x[0] ** (lamb_carvao - 1)) + (1 / (r - drift_p + drift_carvao)) # segunda equação
    return y
```

## Calcular lambdas

### Usina Eólica

```python
lamb0 = -1.0
lamb_eolica_fixo = fsolve(edo_eolica, lamb0)
lamb_eolica_fixo = float(lamb_eolica_fixo)

print(f'O lambda para a usina eólica é {lamb_eolica_fixo}.')
```

### Usina Solar

```python
lamb0 = -1.0
lamb_solar_fixo = fsolve(edo_solar, lamb0)
lamb_solar_fixo = float(lamb_solar_fixo)

print(f'O lambda para a usina solar é {float(lamb_solar_fixo)}.')
```

### Usina a Gás

```python
lamb0 = -1.0
lamb_gas_fixo = fsolve(edo_gas, lamb0)
lamb_gas_fixo = float(lamb_gas_fixo)

print(f'O lambda para a usina a gás natural é {float(lamb_gas_fixo)}.')
```

### Usina a Carvão

```python
lamb0 = -1.0
lamb_carvao_fixo = fsolve(edo_carvao, lamb0)
lamb_carvao_fixo = float(lamb_carvao_fixo)

print(f'O lambda para a usina a carvão é {float(lamb_carvao_fixo)}.')
```

```python
par_solar = pd.DataFrame({})

i_solar = 183.641
r = 0.03
sigma_p_simul = np.arange(.05, .26, 0.01)

for s in sigma_p_simul:
#     print(i_solar)
    x0 = 1
    
    def edo_solar_simul(lamb):
        return (1/2)*((s ** 2) - (2 * rho_solar * s * sigma_solar) + (sigma_solar ** 2)) * lamb * (lamb-1) + (drift_p - drift_solar) * lamb - r 
    
    lamb0 = -1.0
    lamb_solar = root(edo_solar_simul, lamb0)
    lamb_solar = float(lamb_solar.x)
#    print(lamb_solar)
    
    B_solar = 2000
    
    def sistema_solar(x): # x = (x, B)
        y = B_solar * (x[0] ** lamb_solar) + (x[0] / (r - drift_p + drift_solar)) - (1 / r) - i_solar
        return y
    
    def VPL_rc_solar(i):
        x = ((1 / r) + i) * (r - drift_p + drift_solar)
        return x
    
    x_vpl_solar = VPL_rc_solar(i_solar)
    
    solve_solar = root(sistema_solar, x0)
    x_solar = float(solve_solar.x[0])
    
    i_x = {'sigma_p_simul': s, 'x_solar': x_solar, 'x_vpl_solar': x_vpl_solar, 'B_solar': B_solar}
    par_solar =  par_solar.append(i_x, ignore_index = True)
   
par_solar
```

```python
plt.figure(figsize=(9,6))
plt.plot(par_solar['sigma_p_simul'], par_solar['x_solar'], color = 'b')
plt.plot(par_solar['sigma_p_simul'], par_solar['x_vpl_solar'], color = 'black', linestyle='--')
plt.title('Usina Solar - Simulação para sigma_p')
plt.xlabel('Desvio Padrão do Preço da Energia Elétrica')
plt.ylabel('Razão Crítica X = P/C')

custom_lines = [Line2D([0], [0], color='b'),
                Line2D([0], [0], linestyle='--', color='black')]

plt.legend(custom_lines, ['X crítico (OR)', 'X crítico (VPL)'])

plt.savefig('grafs/sim_sigma_solar.png', bbox_inches='tight')
plt.show()
```
